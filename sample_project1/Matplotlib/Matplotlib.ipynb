# Matplotlib example workflow 

```python
#0. import matplotlib and get it ready for plotting in Jupyter 
%matplotlib inline 
import matplotlib.pyplot as plt 
plt.close('all')  # remove any previously stored plots

# 1. Prepare data 
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Set up plot, use object-oriented method
fig, ax = plt.subplots(figsize=(10, 10))  # parameters in figsize indicate height and width

# 3. Plot the data 
ax.plot(x, y)

# 4. Customize plot 
ax.set(title="Simple plot", 
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save and show (save the whole figure)
fig.savefig('sample-plot.png')

# 6. Show the plot
plt.show()

# 7. Close the plot
plt.close('fig')
```

## Making figures with NumPy arrays 

```python
import numpy as np
# Create data
x = np.linspace(0, 10, 100)
print(f'first 10 elements of x = \n{x[:10]}')

plt.close('all')

# Plot the data 
fig, ax = plt.subplots()
ax.plot(x, x**2)
plt.show()
plt.close('fig')
```

```python
# Use the same data to create a scatter plot 
plt.close('all')

fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
plt.show()
plt.close('fig')
```

```python
# Another scatter plot 
plt.close('all')
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
plt.show()
plt.close('fig')
```

```python
# Make a plot from a dictionary 
plt.close('all')
nut_butter_prices = {'Almond Butter': 10,
                     'Peanut Butter': 8,
                     'Cashew Butter': 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), 
       height=nut_butter_prices.values())
ax.set(title="Nacho's Nutter Butter store ", 
      ylabel="Price ($)")
plt.show()
plt.close('fig')
```

# Matplotlib example workflow 

```python
#0. import matplotlib and get it ready for plotting in Jupyter 
%matplotlib inline 
import matplotlib.pyplot as plt 
plt.close('all')  # remove any previously stored plots

# 1. Prepare data 
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Set up plot, use object-oriented method
fig, ax = plt.subplots(figsize=(10, 10))  # parameters in figsize indicate height and width

# 3. Plot the data 
ax.plot(x, y)

# 4. Customize plot 
ax.set(title="Simple plot", 
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save and show (save the whole figure)
fig.savefig('sample-plot.png')

# 6. Show the plot
plt.show()

# 7. Close the plot
plt.close('fig')
```

### Two options for subplots

```python
plt.close('all')
# Subplot option1 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2,
                                             figsize=(10, 5))

# Plot to each different axes
ax1.plot(x, x/2);  # linear function
ax2.scatter(np.random.random(10), np.random.random(10));  # scatter plot
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
plt.show()
plt.close('fig')
```

```python
# Subplots option 2 
plt.close('all')
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.random(1000));
plt.show()
plt.close('fig')
```

### Which one should you use? (pyplot vs matplotlib OO method)

Documentation will push you to do the OO method always
* When plotting something quickly okay to use pyplot method.
* When plotting something more advanced, use the OO method

```python
import matplotlib.pyplot as plt
import numpy as np

def set_dark_style(ax):
    ax.set_facecolor('#2E2E2E')  # Dark grey background
    ax.tick_params(colors='white')  # White tick labels
    ax.xaxis.label.set_color('white')
    ax.yaxis.label.set_color('white')
    ax.title.set_color('white')

plt.close('all')
# Subplot of chol, age, and thalac 
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, ncols=1, figsize=(15, 15))

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='coolwarm')  # Changed colormap for better visibility
# Customize ax0 
ax0.set(title='Heart Disease and Cholesterol Levels',
        xlabel='Age',
        ylabel='Cholesterol levels')
# Add a legend to ax0
legend1 = ax0.legend(*scatter.legend_elements(), title='Target')
plt.setp(legend1.get_title(), color='white')  # Set legend title color
# Add a horizontal line to ax0
ax0.axhline(y=over_50['chol'].mean(), 
            linestyle='--',
            color='yellow',  # Changed color for better visibility
            label='Mean Cholesterol')
legend2 = ax0.legend(loc='lower right')
plt.setp(legend2.get_texts(), color='white')  # Set legend text color

# Add data to ax1
scatter2 = ax1.scatter(x=over_50['age'],
                       y=over_50['thalach'],
                       c=over_50['target'],
                       cmap='coolwarm')  # Changed colormap for better visibility
# Customize ax1
ax1.set(title='Heart Disease and Max Heart Rate',
        xlabel='Age',
        ylabel='Max Heart Rate (bpm)')
# Add a legend ax1
legend3 = ax1.legend(*scatter2.legend_elements(), title='Target')
plt.setp(legend3.get_title(), color='white')  # Set legend title color
# Add horizontal line to ax1
ax1.axhline(y=over_50['thalach'].mean(), 
            linestyle='--', 
            color='yellow',  # Changed color for better visibility
            label='Mean Max Heart Rate')
legend4 = ax1.legend(loc='lower right')
plt.setp(legend4.get_texts(), color='white')  # Set legend text color

# Add data to ax2
bins = np.linspace(over_50['age'].min(), over_50['age'].max(), 20)
for target, color in zip([0, 1], ['#FFA07A', '#98FB98']):  # Light salmon and pale green
    ax2.hist(over_50[over_50['target'] == target]['age'],
             bins=bins,
             alpha=0.7,  # Increased alpha for better visibility
             color=color,
             label=f'Target {target}')
# Customize ax2
ax2.set(title='Age Distribution by Heart Disease Target',
        xlabel='Age',
        ylabel='Count')  # Changed to 'Count' as it's a histogram
# Add a legend to ax2 
legend5 = ax2.legend(title='Target')
plt.setp(legend5.get_title(), color='white')  # Set legend title color
plt.setp(legend5.get_texts(), color='white')  # Set legend text color
# Add vertical line 
ax2.axvline(x=over_50['age'].mean(),
            linestyle='--',
            color='yellow',  # Changed color for better visibility
            label='Mean Age')

# Apply dark style to all subplots
for ax in [ax0, ax1, ax2]:
    set_dark_style(ax)

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold', color='white')
fig.patch.set_facecolor('#1E1E1E')  # Set figure background to dark

plt.tight_layout()
plt.show()
plt.close('fig')
```

```python
import matplotlib.pyplot as plt

def set_dark_style(ax):
    ax.set_facecolor('#2E2E2E')  # Dark grey background
    ax.tick_params(colors='white')  # White tick labels
    ax.xaxis.label.set_color('white')
    ax.yaxis.label.set_color('white')
    ax.title.set_color('white')
    # Set the figure background to match the axes
    ax.figure.patch.set_facecolor('#2E2E2E')

# Close any existing plots
plt.close('all')

# Create a new figure and axis
fig, ax = plt.subplots()

# Plot your data
car_sales['Price'].plot(ax=ax)

# Apply the dark style to the axis
set_dark_style(ax)

# Show the plot
plt.show()

# Close the figure
plt.close(fig)
```

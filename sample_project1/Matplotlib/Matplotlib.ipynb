# Matplotlib example workflow 

```python
#0. import matplotlib and get it ready for plotting in Jupyter 
%matplotlib inline 
import matplotlib.pyplot as plt 
plt.close('all')  # remove any previously stored plots

# 1. Prepare data 
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Set up plot, use object-oriented method
fig, ax = plt.subplots(figsize=(10, 10))  # parameters in figsize indicate height and width

# 3. Plot the data 
ax.plot(x, y)

# 4. Customize plot 
ax.set(title="Simple plot", 
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save and show (save the whole figure)
fig.savefig('sample-plot.png')

# 6. Show the plot
plt.show()

# 7. Close the plot
plt.close('fig')
```

## Making figures with NumPy arrays 

```python
import numpy as np
# Create data
x = np.linspace(0, 10, 100)
print(f'first 10 elements of x = \n{x[:10]}')

plt.close('all')

# Plot the data 
fig, ax = plt.subplots()
ax.plot(x, x**2)
plt.show()
plt.close('fig')
```

```python
# Use the same data to create a scatter plot 
plt.close('all')

fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
plt.show()
plt.close('fig')
```

```python
# Another scatter plot 
plt.close('all')
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
plt.show()
plt.close('fig')
```

```python
# Make a plot from a dictionary 
plt.close('all')
nut_butter_prices = {'Almond Butter': 10,
                     'Peanut Butter': 8,
                     'Cashew Butter': 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), 
       height=nut_butter_prices.values())
ax.set(title="Nacho's Nutter Butter store ", 
      ylabel="Price ($)")
plt.show()
plt.close('fig')
```

```python
# Make a 2nd plot from a dictionary 
plt.close('all')
video_game_cost = {'Halo MCC': 40,
                   'Borderlands': 30,
                   'Borderlands 2': 35}

fig, ax = plt.subplots() 
ax.bar(video_game_cost.keys(), height=video_game_cost.values())
ax.set(title='Video game cost',
      ylabel='Price ($)')
plt.show()
plt.close('fig')
```

```python
plt.close('all')
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), nut_butter_prices.values())
plt.show()
plt.close('fig')
```

```python
# make some data 
x = np.random.randn(1000)  # make 10 random samples
plt.close('all')
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
plt.close('fig')
```

### Two options for subplots

```python
plt.close('all')
# Subplot option1 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2,
                                             figsize=(10, 5))

# Plot to each different axes
ax1.plot(x, x/2);  # linear function
ax2.scatter(np.random.random(10), np.random.random(10));  # scatter plot
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
plt.show()
plt.close('fig')
```

```python
# Subplots option 2 
plt.close('all')
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.random(1000));
plt.show()
plt.close('fig')
```

### Plotting from pandas DataFrames

```python
import pandas as pd 

# Make a dataframe 
car_sales = pd.read_csv('../Pandas/car-sales.csv')
car_sales
```

```python
plt.close('all')
ts = pd.Series(np.random.randn(1000),
              index=pd.date_range('1/1/2020', periods=1000))


ts = ts.cumsum()
ts.plot()
plt.show()
plt.close('ts')
```

```python
car_sales['Price'] = car_sales['Price'].str.replace(r'[\$\,\.]', '')
car_sales
```

```python
# Show that the price column is a string datatype 
type(car_sales['Price'][0])  # This will access the first element of the Price column
```

```python
# Remove the last two zeros 
car_sales['Price'] = car_sales['Price'].str[:-2]
car_sales
```

```python
car_sales['Sales Date'] = pd.date_range('1/1/2020', periods=len(car_sales))
car_sales
```

```python
car_sales['Total Sales'] = car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
# Let's plot the total sales 
plt.close('all')
car_sales.plot(x="Sales Date", y="Total Sales");
plt.show()
plt.close('fig')
```

```python
plt.close('all')  # close any previous graph stored
car_sales['Price'] = car_sales['Price'].astype(int)  # convert the price column into an integer column
car_sales.plot(x='Odometer (KM)', y='Price', kind='scatter') # Show a scatter plot
plt.show()  # call this to show the graph
plt.close('fig')  # Remember to clear the figure 
```

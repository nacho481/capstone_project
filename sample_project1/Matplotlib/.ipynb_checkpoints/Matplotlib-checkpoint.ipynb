# Matplotlib example workflow 

```python
#0. import matplotlib and get it ready for plotting in Jupyter 
%matplotlib inline 
import matplotlib.pyplot as plt 
plt.close('all')  # remove any previously stored plots

# 1. Prepare data 
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Set up plot, use object-oriented method
fig, ax = plt.subplots(figsize=(10, 10))  # parameters in figsize indicate height and width

# 3. Plot the data 
ax.plot(x, y)

# 4. Customize plot 
ax.set(title="Simple plot", 
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save and show (save the whole figure)
fig.savefig('sample-plot.png')

# 6. Show the plot
plt.show()

# 7. Close the plot
plt.close('fig')
```

```python
# 0. import matplotlib and get it read for plotting in Jupyter 
%matplotlib inline 
import matplotlib.pyplot as plt 
plt.close('all')

# 1. Prepare the data 
x_2 = np.random.randint(0, 5, size=5)
y_2 = np.random.randint(0, 5, size=5)

# 2. Set up plot, use object-oriented method 
fig2, ax2 = plt.subplots(figsize=(10, 10))

# 3. Plot the data 
ax2.plot(x_2, y_2)

# 4. Customize plot 
ax2.set(title="2nd graph",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save and show (save the whole figure)
fig.savefig('sample-plot2.png')

# 6. Show the plot
print(x_2)
print(y_2)
plt.show()

# 7. Close the plot
plt.close('fig')
```

## Making figures with NumPy arrays 

```python
import numpy as np
# Create data
x = np.linspace(0, 10, 100)
print(f'first 10 elements of x = \n{x[:10]}')

plt.close('all')

# Plot the data 
fig, ax = plt.subplots()
ax.plot(x, x**2)
plt.show()
plt.close('fig')
```

```python
# Use the same data to create a scatter plot 
plt.close('all')

fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
plt.show()
plt.close('fig')
```

```python
# Another scatter plot 
plt.close('all')
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
plt.show()
plt.close('fig')
```

```python
# Make a plot from a dictionary 
plt.close('all')
nut_butter_prices = {'Almond Butter': 10,
                     'Peanut Butter': 8,
                     'Cashew Butter': 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), 
       height=nut_butter_prices.values())
ax.set(title="Nacho's Nutter Butter store ", 
      ylabel="Price ($)")
plt.show()
plt.close('fig')
```

```python
# Make a 2nd plot from a dictionary 
plt.close('all')
video_game_cost = {'Halo MCC': 40,
                   'Borderlands': 30,
                   'Borderlands 2': 35}

fig, ax = plt.subplots() 
ax.bar(video_game_cost.keys(), height=video_game_cost.values())
ax.set(title='Video game cost',
      ylabel='Price ($)')
plt.show()
plt.close('fig')
```

```python
plt.close('all')
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), nut_butter_prices.values())
plt.show()
plt.close('fig')
```

```python
# make some data 
x = np.random.randn(1000)  # make 10 random samples
plt.close('all')
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
plt.close('fig')
```

### Two options for subplots

```python
plt.close('all')
# Subplot option1 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2,
                                             figsize=(10, 5))

# Plot to each different axes
ax1.plot(x, x/2);  # linear function
ax2.scatter(np.random.random(10), np.random.random(10));  # scatter plot
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
plt.show()
plt.close('fig')
```

```python
# Subplots option 2 
plt.close('all')
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.random(1000));
plt.show()
plt.close('fig')
```

```python
# Let's try another data set 
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease.head()
```

```python
# Convert sex to categorical
heart_disease['sex'] = heart_disease['sex'].map({0: 'Female', 1: 'Male'})
plt.close('all')

# Create a box plot
plt.figure(figsize=(10,6))
heart_disease.boxplot(column='chol', by='sex')
plt.title('Cholesterol by sex')
plt.suptitle('') # This removes automatic suptitle
plt.xlabel('sex')
plt.ylabel('Cholesterol')
plt.show()

# Create histograms
plt.figure(figsize=(12,6))
plt.hist([heart_disease[heart_disease['sex'] == 'Female']['chol'],
         heart_disease[heart_disease['sex'] == 'Male']['chol']],
         label=['Female', 'Male'], bins=30, alpha=0.7)
plt.title('Distribution of Cholesterol Levels by Sex')
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Calculate and print summary statistics
summary = heart_disease.groupby('sex')['chol'].describe()
print(summary)
```

```python
# Create a histogram of age column
plt.close('all')
heart_disease['age'].plot.hist(bins=20)
plt.show()
plt.close('fig')
```

```python
# Create a histogram of age column
plt.close('all')
heart_disease['age'].plot.hist(bins=50)
plt.show()
plt.close('fig')
```

```python
plt.close('all')
heart_disease.plot.hist(figsize=(10, 30), subplots=True);
plt.show()
plt.close('fig')
```

```python
plt.close('all')
chol_by_sex = heart_disease.groupby('chol')['sex']
chol_by_sex.plot(kind='bar')
plt.show()
plt.close('fig')
```

### Which one should you use? (pyplot vs matplotlib OO method)

Documentation will push you to do the OO method always
* When plotting something quickly okay to use pyplot method.
* When plotting something more advanced, use the OO method 

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
print(len(over_50))
```

```python
plt.close('all')
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
plt.show()
plt.close('fig')
```

```python
# OO method
plt.close('all')
fig, ax = plt.subplots(figsize=(10, 6))  # figsize should be inches 
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax); 
plt.show()
```

```python
# OO method
plt.close('all')
fig, ax = plt.subplots(figsize=(10, 6))  # figsize should be inches 
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax,
            figsize=(10, 6)); 
plt.show()
```

```python
# OO method with pyplot method
plt.close('all')
fig, ax = plt.subplots(figsize=(10, 6))  # figsize should be inches 
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax); 
ax.set_xlim(45, 100)
plt.show()
```

```python
plt.close('all')
## OO method form scratch 
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data 
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target']);

# Customize the set 
ax.set(title='Heart Disease and Cholesterol levels',
       xlabel='Age',
       ylabel='Cholesterol');

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target');

plt.show()
plt.close('fig')
```
